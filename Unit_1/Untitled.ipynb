## Queue and Stack

```python
import sys

class Solution:
    
    def __init__(self):
        self.queue = []
        self.stack = []
    
    def pushCharacter(self, i):
        self.stack.append(i)
    
    def popCharacter(self):
        return self.stack.pop(-1)
    
    def enqueueCharacter(self, i):
        self.queue.append(i)
    
    def dequeueCharacter(self):
        return self.queue.pop(0)   

def main():
    s=input("Type a word: ")
    obj=Solution()
    l=len(s)
    for i in range(l):
        obj.pushCharacter(s[i])
        obj.enqueueCharacter(s[i])
        
#     import pdb; pdb.set_trace()
    
    isPalindrome=True
    
    for i in range(l // 2):
        if obj.popCharacter()!=obj.dequeueCharacter():
            isPalindrome=False
            break
            
    if isPalindrome:
        print("The word, "+s+", is a palindrome.")
    else:
        print("The word, "+s+", is not a palindrome.")  
```

```python
if __name__=="__main__":
    main()
```

## Bubble Sort

```python
import sys

def input_data():
    n = int(input().strip())
    a = list(map(int, input().strip().split(' ')))
    
    return n, a

def sort_bubble(n, a):
    nofswap = 0

    for i in range(0,n-1):
        for i in range(0,n-1):
            if a[i]>a[i+1]:
                a[i],a[i+1] = a[i+1],a[i]
                nofswap +=1
                
    return a, nofswap

def main():
    n, a = input_data()
    n, nofswap = sort_bubble(n, a)
    print("Array is sorted in", nofswap, "swaps.")
    print("First Element:",a[0])
    print("Last Element:",a[-1])
```

```python
if __name__=="__main__":
    main()
```

## 2D Arrays

```python
def input_arr():
    arr = []
    n = int(input("Type array size: "))
    for _ in range(n):
        arr.append(list(map(int, input().rstrip().split())))
            
    return arr

def check_hourglass(arr):
    res = []
    for x in range(0, 4):
        for y in range(0, 4):
            s = sum(arr[x][y:y+3]) + arr[x+1][y+1] + sum(arr[x+2][y:y+3])
            res.append(s)
    
    return res
```

```python
if __name__ == '__main__':
    arr = input_arr()
    print(f'Max of sum of hourglass elements: {max(check_hourglass(arr))}')
```
