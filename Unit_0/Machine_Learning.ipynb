## Sums

$\sum\limits_{i=0}^Nn$

```python
def summ(n, i, N):
    """Return the sum of n from i to N"""
    return (N-i)*n
```

```python
def summm(n, i, N):
    temp = 0
    for _ in range(i, N+1):
        temp = temp + n
    
    return temp
```

```python
n = 1
i = 0
N = 2
print(f"{summm(n, i, N)}")
# Returns 4 >>> N-1
```

```python
n = 1
i = 1
N = 5
print(f"{summ(n, i, N)}")
# Returns 4 >>> N-1
```

$\sum\limits_{t=1}^T\sum\limits_{k=1}^Kn$

```python
def double_summ(n, k, K, t, T):
    temp1 = 0
    temp2 = 0
    for _ in range(k, K+1):
        for _ in range(t, T+1):
            temp1 = temp1 + n
        temp2 = temp2 + temp1
    
    return temp2
```

```python
def summ_pow(n, low_lim, up_lim):
    temp = 0
    for i in range(low_lim, up_lim+1):
        temp = temp + n**i
    
    return temp
```

```python
def multiply_vector(x, y):
    """Multiply each elements of the vectors"""
    assert len(x)==len(y), "Vectors must be the same length"
    temp = [i*j for i, j in zip(x, y)]
    
    return sum(temp)

def get_angle(x, y):
    """Returns the angle between vectors"""
    cos_angle = multiply_vector(x, y)/(vector_length(x)*vector_length(y))
    
    return f"{np.arccos(cos_angle):.4f}"
```

```python
print(f"Angle between x and y: {get_angle(x, y)} rad")
```

```python
def get_sum_metrics(predictions, metrics=None):

    if metrics == None :
        metrics = []

    for i in range(3):
        metrics.append(lambda x, i=i: x + i)

    sum_metrics = 0
    for metric in metrics:
        sum_metrics += metric(predictions)

    return sum_metrics


def main():
    print(get_sum_metrics(0))  # Should be (0 + 0) + (0 + 1) + (0 + 2) = 3
    print(get_sum_metrics(1))  # Should be (1 + 0) + (1 + 1) + (1 + 2) = 6
    print(get_sum_metrics(2))  # Should be (2 + 0) + (2 + 1) + (2 + 2) = 9
    print(get_sum_metrics(3, [lambda x: x]))  # Should be (3) + (3 + 0) + (3 + 1) + (3 + 2) = 15
    print(get_sum_metrics(0))  # Should be (0 + 0) + (0 + 1) + (0 + 2) = 3
    print(get_sum_metrics(1))  # Should be (1 + 0) + (1 + 1) + (1 + 2) = 6
    print(get_sum_metrics(2))  # Should be (2 + 0) + (2 + 1) + (2 + 2) = 9

if __name__ == "__main__":
    main()
```

```python
import pandas as pd

movies = {
    'comedy': [0, 1, 0, 1],
    'action': [1, 1, 1, 1],
    'spielberg': [0, 1, 0, 0],
    'dino': [0, 0, 1, 1],
    'disney': [1, 0, 1, 0],
    'like/dislike': [1,-1, 1, 1]
}

df = pd.DataFrame(movies,
                  index=[
                      'movie 1','movie 2','movie 3','movie 4' 
                  ])
display(df)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import numpy as np

seq = 'like/dislike'

x = df.drop([seq], axis=1)
y = df[seq].to_list()
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1) 
```

```python
classifier = LogisticRegression(solver='lbfgs',random_state=0)
classifier.fit(x_train, y_train)
print('Accuracy: {:.2f}'.format(classifier.score(x_test, y_test)))
```

$h(x;\theta)=sign(\theta\cdot x)$

```python
def dot(a, b):
    return sum([i*j for i,j in zip(a,b)])

theta = [1,-1]
x = [[1,-1],[1,0],[0,1],[0,0]]

for i in x:
    print(f"sign {theta}.{i} = {dot(theta, i)}.")
```

$X:x\cdot\theta=0$<br>
$\text{if } x\cdot\theta>0\text{ then } X \text{ is positive}$

```python
theta_0 = -1

for i in x:
    print(f"sign {theta}.{i} = {dot(theta, i)+theta_0}.")
```

$X:x\cdot\theta - \theta_0=0$<br>
$\text{if } x\cdot\theta- \theta_0>0\text{ then } X \text{ is positive}$

Linear separation<br>
Example: $h: X\rightarrow \{-1,0,1\}$
