## Sums

$\sum\limits_{i=0}^Nn$

```python
def summ(n, i, N):
    """Return the sum of n from i to N"""
    return (N-i)*n
```

```python
def summm(n, i, N):
    temp = 0
    for _ in range(i, N+1):
        temp = temp + n
    
    return temp
```

```python
n = 1
i = 0
N = 2
print(f"{summm(n, i, N)}")
# Returns 4 >>> N-1
```

```python
n = 1
i = 1
N = 5
print(f"{summ(n, i, N)}")
# Returns 4 >>> N-1
```

$\sum\limits_{t=1}^T\sum\limits_{k=1}^Kn$

```python
def double_summ(n, k, K, t, T):
    temp1 = 0
    temp2 = 0
    for _ in range(k, K+1):
        for _ in range(t, T+1):
            temp1 = temp1 + n
        temp2 = temp2 + temp1
    
    return temp2
```

```python
n = 1
K = 4
k = 1
T = 5
t = 1
```

```python
print(summ(summ(n, k, K),t ,T))
print((T-t)*(K-k)*1)
```

```python
print(double_summ(n, k, K, t, T))
```

$\sum\limits_{t=1}^T\sum\limits_{k=1}^K0.5^t$

```python
def summ_pow(n, low_lim, up_lim):
    temp = 0
    for i in range(low_lim, up_lim+1):
        temp = temp + n**i
    
    return temp
```

```python
print(summ_pow(0.5, 1, 100000))
```

```python
def vector_length(x):
    temp = 0
    for i in x:
        temp = temp+i**2 
        
    return temp**.5

x = (1.2, 1.5)
print((1.2**2+1.5**2)**.5)
print(vector_length(x))
```

```python
x = (.4, .3)
print(f"x = {x}")
print(f"length x = {vector_length(x)}")
```

```python
y = (-.15, .2)
print(f"y = {y}")
print(f"length {y} = {vector_length(y)}")
```

```python
def multiply_vector(x, y):
    """Multiply each elements of the vectors"""
    assert len(x)==len(y), "Vectors must be the same length"
    temp = [i*j for i, j in zip(x, y)]
    
    return sum(temp)

def get_angle(x, y):
    """Returns the angle between vectors"""
    cos_angle = multiply_vector(x, y)/(vector_length(x)*vector_length(y))
    
    return f"{np.arccos(cos_angle):.4f}"
```

```python
print(f"Angle between x and y: {get_angle(x, y)} rad")
```

```python
def get_sum_metrics(predictions, metrics=[]):
    for i in range(3):
        metrics.append(lambda x: x + i)

    sum_metrics = 0
    for metric in metrics:
        sum_metrics += metric(predictions)

    return sum_metrics


def main():
    print(get_sum_metrics(0))  # Should be (0 + 0) + (0 + 1) + (0 + 2) = 3
    print(get_sum_metrics(1))  # Should be (1 + 0) + (1 + 1) + (1 + 2) = 6
    print(get_sum_metrics(2))  # Should be (2 + 0) + (2 + 1) + (2 + 2) = 9
    print(get_sum_metrics(3, [lambda x: x]))  # Should be (3) + (3 + 0) + (3 + 1) + (3 + 2) = 15
    print(get_sum_metrics(0))  # Should be (0 + 0) + (0 + 1) + (0 + 2) = 3
    print(get_sum_metrics(1))  # Should be (1 + 0) + (1 + 1) + (1 + 2) = 6
    print(get_sum_metrics(2))  # Should be (2 + 0) + (2 + 1) + (2 + 2) = 9

if __name__ == "__main__":
    main()
```
